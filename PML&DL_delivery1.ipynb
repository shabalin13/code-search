```python
!pip install transformers --quiet


from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("microsoft/codebert-base")

model = AutoModel.from_pretrained("microsoft/codebert-base")
```

```python
from transformers import AutoTokenizer, AutoModel
import torch


def get_emb(code='', nl_text=''):
    nl_tokens=tokenizer.tokenize(nl_text)
    code_tokens=tokenizer.tokenize(code)
    tokens=[tokenizer.cls_token]+nl_tokens+[tokenizer.sep_token]+code_tokens+[tokenizer.sep_token]
    tokens_ids=tokenizer.convert_tokens_to_ids(tokens)
    t = torch.tensor(tokens_ids).expand(1, -1)
    context_embeddings, embedding = model(t)[0:2]
    return embedding


code1 = "def max(a,b): if a>b: return a else return b"
code2 = "def max(b,c): if b>c: return b else return c"
code3 = "def min(a,b): if a<b: return a else return b"
code4 = "def sub(a,b): return a-b"
code5 = 'hadoop123h'


def measure_distance(text1, text2, emb_func):
  print('Distance between:')
  print(text1)
  print(text2)
  print('is: ' + str(torch.cdist(emb_func(text1), emb_func(text2))))
  print()
  

measure_distance(code1, code2, get_emb)
measure_distance(code1, code3, get_emb)
measure_distance(code1, code4, get_emb)
measure_distance(code2, code4, get_emb)

measure_distance(code1, code5, get_emb)

nl_text = 'return maximum value'

measure_distance(code1, nl_text, get_emb)
measure_distance(code2, nl_text, get_emb)
measure_distance(code3, nl_text, get_emb)
measure_distance(code4, nl_text, get_emb)
measure_distance(code5, nl_text, get_emb)
```
